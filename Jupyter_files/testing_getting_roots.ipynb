```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([0, 1, 2, 3, 4, 5, 6, 7])
y = np.array([ 9,  4,  0, -1, -1,  4,  8, 16])

poly = np.polyfit(x,y,2)
y0 = 4

p = np.poly1d(poly)

x0 = (p - y0).roots


plt.plot(x,y)
plt.plot(np.full(8,y0))
plt.plot(np.linspace(0,8,50), p(np.linspace(0,8,50)))

plt.show()

print(x0)
```
